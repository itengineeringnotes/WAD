```python
import pygame
import sys
```

```python
pygame.init()
```

```python
# Window settings
WIDTH, HEIGHT = 300, 300
GRID = 3
CELL = WIDTH // GRID
```

```python
# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
```

```python
# Screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Tic Tac Toe")
```

```python
# Board
board = [[' ']*GRID for _ in range(GRID)]
```

```python
# Draw grid
def draw_grid():
    for i in range(1, GRID):
        pygame.draw.line(screen, BLACK, (i*CELL, 0), (i*CELL, HEIGHT), 2)
        pygame.draw.line(screen, BLACK, (0, i*CELL), (WIDTH, i*CELL), 2)
```

```python
# Draw X or O
def draw_symbols():
    font = pygame.font.Font(None, 100)
    for r in range(GRID):
        for c in range(GRID):
            if board[r][c] != ' ':
                text = font.render(board[r][c], True, BLACK)
                rect = text.get_rect(center=(c*CELL + CELL//2, r*CELL + CELL//2))
                screen.blit(text, rect)
```

```python
#Check winner
def check(symbol):
    for i in range(GRID):
        if all(board[i][j] == symbol for j in range(GRID)) or \
           all(board[j][i] == symbol for j in range(GRID)):
            return True

    return all(board[i][i] == symbol for i in range(GRID)) or \
           all(board[i][GRID-1-i] == symbol for i in range(GRID))
```

```python
# Check draw
def full():
    return all(board[r][c] != ' ' for r in range(GRID) for c in range(GRID))
```

```python
# Reset game
def reset():
    global board
    board = [[' ']*GRID for _ in range(GRID)]
```

```python
turn = 'X'
running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.MOUSEBUTTONDOWN:
            x, y = event.pos
            row, col = y // CELL, x // CELL

            if board[row][col] == ' ':
                board[row][col] = turn

                if check(turn):
                    print(turn, "wins!")
                    reset()

                elif full():
                    print("Draw!")
                    reset()

                turn = 'O' if turn == 'X' else 'X'

    screen.fill(WHITE)
    draw_grid()
    draw_symbols()
    pygame.display.update()

pygame.quit()
sys.exit()
```
